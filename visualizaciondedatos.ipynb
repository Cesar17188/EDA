```python
#Llamamos el set de datos público
zoo_data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data'

# Los nombres de las columnas se encuentran en : 'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.names
zoo_name = ['animal name',
            'hair',
            'feathers',
            'eggs',
            'milk',
            'airborne',
            'aquatic',
            'predator',
            'toothed',
            'backbone',
            'breathes',
            'venomous',
            'fins',
            'legs',
            'tail',
            'domestic',
            'cat-size',
            'type']
```

```python
import pandas as pd
```

```python
df_zoo = pd.read_csv(zoo_data)
```

```python
df_zoo.head(10)
```

```python
df_zoo.columns = zoo_name
```

```python
df_zoo.head()
```

```python
import altair as alt
```

```python
plot_01 = alt.Chart(df_zoo).mark_line().encode(
    x = 'animal name',
    y = 'legs'
)
```

```python
plot_01
```

```python
plot_02=alt.Chart(df_zoo).mark_bar().encode(
    x='animal name',
    y='legs'
)
```

```python
plot_02
```

```python
import plotly.express as px
```

```python
px.pie(df_zoo,values='aquatic',names='animal name', title='Grafica de pie ')
```
